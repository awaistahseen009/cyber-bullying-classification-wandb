```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('cyberbullying_tweets.csv')
```

```python
df.head()
```

```python
df.columns= ['text','target']
```

```python
df['target'].unique()
```

```python
df.isna().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.pie(df['target'].value_counts(),labels=df['target'].value_counts().index,autopct="%.1f%%")
plt.title('Categories Percentage')
plt.show()
```

### Result 1: As all categories have equal percentage so our data is not imbalanced

# Countplot to see the count of values

```python
sns.countplot(x='target', data=df)
plt.xticks(rotation='vertical')  # Rotate x-axis labels vertically
plt.show()
```

```python
from wordcloud import WordCloud
```

```python
for cat in df['target'].unique():
    genre_descriptions = ' '.join(df[df['target'] == cat]['text'])
    wordcloud = WordCloud(width=500, height=500, background_color='black').generate(genre_descriptions)
    plt.figure(figsize=(8, 8), facecolor=None)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.title(f'Word Cloud for {cat} Category')
    plt.show()
```
